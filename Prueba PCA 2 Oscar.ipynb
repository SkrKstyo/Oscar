```python
import matplotlib.pyplot as plt

from jupyterthemes import jtplot
jtplot.style()


import pandas as pd
import seaborn as sns
from sympy import *

from sklearn.decomposition import PCA

import numpy as np
```

```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

#read data
data = pd.read_csv('Country-data.csv')
data.head()
```

```python
data.head()
```

```python
X = data[['child_mort', 'exports', 'health', 'imports','income', 'inflation', 'life_expec', 'total_fer','gdpp']]
```

```python
y = data['country']
```

```python
data.head()
```

```python
X
```

```python
pca = PCA()
```

```python
pca
```

```python
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
vals = pca.explained_variance_ratio_
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.bar(np.arange(1, len(vals) + 1), pca.explained_variance_ratio_, color = 'g', alpha = 0.7)
ax.step(np.arange(1, len(vals) + 1), np.cumsum(pca.explained_variance_ratio_), linestyle = '--')
ax.set_xlabel('Componentes')
ax.set_ylabel('Varianza Explicada');
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
newX2 = pca.transform(X)
```

```python
newX2
```

```python
newX2 = np.column_stack([newX2, y])
```

```python
newX2
```

```python
colors = ['magenta','black','Orange','pink','yellow','green']
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
    
for f, c in zip(y.unique(), colors):
    mask = newX2[:, 2] == f
    ax.scatter(newX2[mask, 0], newX2[mask, 1], color = c, label = f)

ax.legend();
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

for f in zip(y.unique()):
    mask = newX2[:, 2] == f
    ax.scatter(newX2[mask, 0], newX2[mask, 1], color = 'b', label = f)
    
```
