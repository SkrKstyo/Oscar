```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su gradiente
def func(x, y):
    return 3 * x**2 + 7 * y**2

def gradient(x, y):
    df_dx = 6 * x
    df_dy = 14 * y
    return np.array([df_dx, df_dy])

# Algoritmo de descenso del gradiente
def gradient_descent(initial_point, learning_rate, num_iterations):
    history = []
    current_point = initial_point.copy()

    for _ in range(num_iterations):
        history.append(current_point)
        grad = gradient(current_point[0], current_point[1])
        current_point = current_point - learning_rate * grad

    return np.array(history)

# Parámetros iniciales
initial_point = np.array([9.0, 10.0])
Rango = 0.1
num_iterations = 3000

# Ejecutar el descenso del gradiente
trajectory = gradient_descent(initial_point, learning_rate, num_iterations)

# Graficar la función y la trayectoria del descenso del gradiente
x_vals = np.linspace(-10, 10, 100)
y_vals = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = func(X, Y)

plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.scatter(trajectory[:, 0], trajectory[:, 1], c='red', label='Descenso del Gradiente')
plt.title('Descenso del Gradiente para 3*x**2 + 7*y**2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
